```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, r2_score
import matplotlib.pyplot as plt
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
rmhd_trn = pd.read_csv('Mohos-train.csv',index_col=0)
rmhd_trn.head(5)
```

```python
rmhd_trn.isnull().sum()
```

```python
X = rmhd_trn.drop(columns='Hardness')
y = rmhd_trn.Hardness
```

```python
dtr = DecisionTreeRegressor(random_state=24)
kfold = KFold(n_splits=5, shuffle=True, random_state=24)
scaler_mm = MinMaxScaler()
scaler_std = StandardScaler()
pipe = Pipeline([('SCL',None),('DTR',dtr)])
params = {'DTR__min_samples_leaf' : [2,10,25,50],
'DTR__min_samples_split' : [1,10,25,50],
'DTR__max_depth' : [None, 2, 4, 6,8,10,12,14,16,18], 'SCL': [scaler_mm,scaler_std,None]}
gcv = GridSearchCV(pipe, param_grid=params, cv = kfold,scoring='r2', verbose=3)
gcv.fit(X,y)
```

```python
print(gcv.best_score_)
print(gcv.best_params_)
```

```python
tst = pd.read_csv('Mohos-test.csv',index_col=0)
y_pred = gcv.predict(tst)
y_pred
```

```python
submit = pd.read_csv('Mohos-sample_submission.csv')
submit.price = y_pred
submit.to_csv('Submission_RMHD_Predictiion.csv', index=False)
```
