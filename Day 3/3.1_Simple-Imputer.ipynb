```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.impute import SimpleImputer
```

```python
chem = pd.read_csv('ChemicalProcess.csv')
chem.head()
```

```python
chem.isnull().sum().sum()             # .isnulll() to chechk total number of null values in dataset
```

# Constant Strategy

```python
#Applying Impoter Comnstant Strategy 
imp_const = SimpleImputer(strategy='constant',fill_value=0.5).set_output(transform='pandas')
imp_const
```

```python
chem_const = imp_const.fit_transform(chem)
chem_const.isnull().sum().sum()
```

# Mean Strategy

```python
#Applying Impoter mean Strategy 
imp_mean = SimpleImputer(strategy='mean',fill_value=0.5).set_output(transform='pandas')
chem_mean = imp_mean.fit_transform(chem)
chem_mean.isnull().sum().sum()
```

- `.fit()`     :  Calculates the mean or median according to strategy. 
- `.transform()` :  calculates does the replacing of missing values wih mean or median according to the strategy.
- `.fit_transform` : Does the work of both of them one by one (first: `.fit()` & then `.transform()`).

```python
imp_mean.statistics_
```

- OR

```python
X = chem.drop('Yield',axis=1)
y = chem['Yield']
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 24, test_size=0.3)
```

```python
# Method 1 : 
imp_mean = SimpleImputer(strategy='mean',fill_value=0.5).set_output(transform='pandas')
imp_mean.fit(X_train)
X_imp_trn = imp_mean.transform(X_train)
X_imp_tst = imp_mean.transform(X_test)
print(X_imp_trn.isnull().sum().sum())
print(X_imp_tst.isnull().sum().sum())
```

# Median Strategy

```python
#Applying Impoter Median Strategy 
imp_med = SimpleImputer(strategy='median',fill_value=0.5).set_output(transform='pandas')
chem_med = imp_med.fit_transform(chem)
chem_med.isnull().sum().sum()
```
