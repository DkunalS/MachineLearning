```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
```

```python
air = pd.read_csv('AirPassengers.csv',index_col=0)
air.plot()
plt.show()
```

```python
air.index = pd.to_datetime(air.index).to_period('M')
```

```python
m_pass = air.resample('Q').sum()
m_pass.index.rename('Quarter', inplace = True)
m_pass.plot()
plt.title('Quarterly Passengers')
plt.show()
```

```python
y_pass = air.resample('Y').sum()
y_pass.index.rename('Yearly', inplace = True)
y_pass.plot()
plt.title('Yearly Passengers')
plt.show()
```

### Comparison to Monthly & Quarterly & Yearly

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6)) 

air.plot(ax=axes[0])
axes[0].set_title('Monthly Passengers')

m_pass = air.resample('Q').sum()
m_pass.index.rename('Quarter', inplace=True)
m_pass.plot(ax=axes[1])
axes[1].set_title('Quarterly Passengers')

y_pass = air.resample('Y').sum()
y_pass.index.rename('Year', inplace=True)
y_pass.plot(ax=axes[2])
axes[2].set_title('Yearly Passengers')

plt.tight_layout()
plt.show()
```
