```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
import numpy as np
from sklearn.compose import make_column_transformer 
from sklearn.compose import make_column_selector
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss, accuracy_score, r2_score
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold, GridSearchCV, KFold
from sklearn.linear_model import LinearRegression, ElasticNet
from sklearn.pipeline import Pipeline
import os 
import warnings
warnings.filterwarnings('ignore')
```

```python
playg = pd.read_csv(r"playground-series-s4e9\train.csv",index_col=0)
playg
```

```python
playg.info()
```

```python
playg.isnull().sum()
```

```python
playg['fuel_type'].value_counts(dropna=False)
```

```python
playg['fuel_type'].value_counts()
```

```python
X_train = playg.drop(columns='price')
y_train = playg.price
```

```python
ohe = OneHotEncoder(handle_unknown='infrequent_if_exist', sparse_output=False, min_frequency=3000).set_output(transform='pandas')
```

```python
ct_ohe = make_column_transformer(('passthrough', make_column_selector(dtype_exclude=object)),
                                (ohe, make_column_selector(dtype_include=object)),
                                verbose_feature_names_out=False).set_output(transform='pandas')
X_train_ohe = ct_ohe.fit_transform(X_train)
X_train_ohe
```

```python
X_train_ohe.isnull().sum().sum()
```

```python
X_test = pd.read_csv(r"playground-series-s4e9\test.csv",index_col=0)
X_test.info()
```

```python
X_test_ohe = ct_ohe.transform(X_test)
```

```python
X_test_ohe
```

# Using Linear Regression

```python
lr = LinearRegression()
lr.fit(X_train_ohe, y_train)
y_pred = lr.predict(X_test_ohe)
```

```python
y_pred[y_pred < 150] = 150
```

```python
np.quantile(y_pred,[0, 0.25, 0.5, 0.75, 1.0])
```

```python
submit = pd.read_csv('playground-series-s4e9/sample_submission.csv')
```

```python
submit.price = y_pred
```

```python
submit.to_csv('playground-series-s4e9/linearRegression_submission.csv', index=False)
```

# Using the Elastic Net


```python
en = ElasticNet()
kfold = KFold(n_splits=5, shuffle=True, random_state=24)
pipe = Pipeline([('CT', ct_ohe), ('EN',en)])

params = {'EN__alpha': np.linspace(0.001, 5, 3), 'EN__l1_ratio' : np.linspace(0.001, 1, 3)}
gcv_en = GridSearchCV(pipe,param_grid=params, cv=kfold, verbose=3)
gcv_en.fit(X_train, y_train)
```

```python
print("Best Param: ", gcv_en.best_params_)
print("Best Score: ", gcv_en.best_score_)
```

```python
bm_el = gcv_en.best_estimator_
y_pred = bm_el.predict(X_test)
y_pred
```

```python
y_pred[y_pred < 150] = 150
```

```python
np.quantile(y_pred,[0, 0.25, 0.5, 0.75, 1.0])
```

```python
submit = pd.read_csv('playground-series-s4e9/sample_submission.csv')
```

```python
submit.price = y_pred
```

```python
submit.to_csv('playground-series-s4e9/elasticNet_submission.csv', index=False)
```

# Using the KNN

```python
knnr = KNeighborsRegressor()
scaler_mm = MinMaxScaler()
scaler_std = StandardScaler()

pipe = Pipeline([('CT',ct_ohe),('SCL',None),('KNNR',knnr)])
pipe.get_params()
```

```python
kfold = KFold(n_splits=5, random_state=24, 
                        shuffle=True)
# here we are checking which KNN value is best in the range
params = {'KNNR__n_neighbors': np.arange(1,11),
         'KNNR__metric':['minkowski'],
          'SCL':[scaler_mm, scaler_std, None]}

gcv = GridSearchCV(pipe, param_grid=params,
                   scoring='r2',          
                   cv=kfold, verbose=3)
gcv.fit(X_train,y_train)
print("Best Parameters: ",gcv.best_params_)
print("Best Score: ",gcv.best_score_)

pd_cv = pd.DataFrame(gcv.cv_results_)
print("Shape: ",pd_cv.shape)
```

```python
bm_knn = gcv.best_estimator_
y_pred = bm_knn.predict(X_test)
y_pred
```

```python
y_pred[y_pred < 150] = 150
```

```python
np.quantile(y_pred,[0, 0.25, 0.5, 0.75, 1.0])
```

```python
submit = pd.read_csv('playground-series-s4e9/sample_submission.csv')
```

```python
submit.price = y_pred
```

```python
submit.to_csv('playground-series-s4e9/knn_submission.csv', index=False)
```
